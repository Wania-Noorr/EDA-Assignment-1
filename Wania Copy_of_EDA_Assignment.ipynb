```python
# Filling the missing values in numerical columns with the median
df.fillna(df.median(numeric_only=True), inplace=True)

# Filling the missing values in categorical columns with the mode
for col in df.select_dtypes(include=['object']):
    df[col].fillna(df[col].mode()[0], inplace=True)

# filling the missing values with the median means that the values will not be easily influenced by extreme values and helps keep the data uniform
# mode is the most repeated value, meaning that the mode value will represent the most common or repeated categories
```

### **Analysis**


 ( Hint: Group by, conditional filtering, Visualization )

##### Which membership type shows the highest total spending in the dataset?  (2 Points)

```python
# Calculating the total spending by membership type
membership_spending = df.groupby('Membership Type')['Total Spend'].sum()

# Finding the membership type with the highest total spending
highest_spending_membership = membership_spending.idxmax()

print(f"The membership type with the highest total spending is: {highest_spending_membership}")
```

#### How does customer satisfaction impact total spending across different membership types? Which membership type spends more based on satisfaction levels? (2 Points)

```python
# Grouping the data by membership type and satisfaction level, then calculating the total spending
spending_by_satisfaction = df.groupby(['Membership Type', 'Satisfaction Level'])['Total Spend'].sum()

# Finding the highest spending membership type for each satisfaction level
highest_spending = spending_by_satisfaction.groupby('Satisfaction Level').idxmax()

print("Membership type with the highest spending for each satisfaction level:")
print(highest_spending)
```

#### What is the average rating for customers who applied discounts, and does the membership type have any effect on the ratings? (2 points)

```python
# Calculating the average rating for customers who applied discounts
avg_rating_with_discount = df[df['Discount Applied'] == True]['Average Rating'].mean()
print(f"Average rating for customers with discounts: {avg_rating_with_discount:.2f}")

# Calculating the average ratings for customers with discounts, grouped by membership type
avg_rating_by_membership = df[df['Discount Applied'] == True].groupby('Membership Type')['Average Rating'].mean()
print("\nAverage rating for customers with discounts by membership type:")
print(avg_rating_by_membership)
```

#### Analyze the Impact of Days Since Last Purchase on Satisfaction Level (2 Points)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group data by 'Days Since Last Purchase' and calculate the average satisfaction level
avg_satisfaction_by_days = df.groupby('Days Since Last Purchase')['Satisfaction Level'].apply(lambda x: (x == 'Satisfied').mean())

# Create a bar plot to visualize the impact
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_satisfaction_by_days.index, y=avg_satisfaction_by_days.values)
plt.xlabel("Days Since Last Purchase")
plt.ylabel("Average Satisfaction Level")
plt.title("Impact of Days Since Last Purchase on Satisfaction Level")
plt.show()
```
